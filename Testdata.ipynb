```python
sales_data = sales_data.loc[:, (sales_data.isnull().sum(axis=0) <= 200)]
```

```python
mid = sales_data['regular_price']
```

```python
sales_data.drop(labels=['regular_price'], axis=1,inplace = True)
```

```python
sales_data.insert(4, 'regular_price', mid)
```

```python
sales_data.drop(['attribute_data:pa_brand'], axis=1, inplace=True)
```
